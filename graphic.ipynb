```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Dados fornecidos
x = ["Parte 1", "Parte 2", "Parte 3", "Parte 4", "Parte 5"]
y = [0.66746, 0.76555, 0.77033, 0.69856, 0.78229]

# Criar o DataFrame
df = pd.DataFrame({'Part': x, 'Value': y})

# Criar o gráfico de linhas com seaborn
plt.figure(figsize=(10, 6))
line_plot = sns.lineplot(x='Part', y='Value', data=df, marker='o')

# Adicionar os valores de Y na linha do gráfico
for i in range(len(df)):
    plt.text(df['Part'][i], df['Value'][i], round(df['Value'][i], 5), 
             horizontalalignment='left', size='medium', color='black', weight='semibold')

# Adicionar título e rótulos aos eixos
plt.title('Evolução do Score Kaggle')
plt.xlabel('Etapas')
plt.ylabel('Escore')

# Mostrar o gráfico
plt.show()
```
